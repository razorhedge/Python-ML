# Carga de Datos a traves de Open

```python
data3 = open(mainpath + "/" + "customer-churn-model/Customer Churn Model.txt", "r")
```

```python
cols = data3.readline().strip().split(",")
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col]=[]
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter +=1

print("El dataset tiene %d filas y %d columnas"%(counter, n_cols))
```

```python
df3= pd.DataFrame(main_dict)
df3.head()
```

# Lectura y Escritura de Ficheros

```python
infile = mainpath + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "/" + "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile) as infile1:
    with open(outfile,"w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

# Leer Datos desde una URL

```python
medals_url = "http://winterolympicsmedals.com/medals.csv"
```

```python
medals_data = pd.read_csv(medals_url, sep=",")
```

```python
medals_data
```

```python
import csv
import urllib3
```

```python
http = urllib3.PoolManager()
req = http.request("GET", medals_url)
req.status
response = req.data
```

```python
cr= csv.reader(response)
cr
```

```python
outfile2 = mainpath + "/" + "URL/Tab Url Text.txt"
with open(outfile2, "w") as url_data:
    for line in cr:
        fields = line.strip().split("\n")
        for i in range(len(fields)):
            fields2 = line.strip().split(",")
            url_data.write("\t".join(fields2))
            url_data.write("\n")
```
